```python
df=pd.read_csv("/content/ai_resume_screening (1).csv")
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
pd.set_option("display.width",None)
pd.set_option("display.max_colwidth",None)
print(df.to_string())
file_path = "data.txt"
content=df.to_string()
with open(file_path,'w') as file:
    file.write(content)
print(f"File '{file_path}' created successfully with 'w' mode.")
```

```python
print(f"Rows-{df.shape[0]} and columns-{df.shape[1]}")
print(f"The columns are \n{df.columns}")
print("The datatypes row-wise are:-\n",df.dtypes)
print(df.head())
print(df.tail())
Y=0
N=0
for i in content.split():
  if i=="Yes":
    Y+=1
  if i=="No":
    N+=1
print(f"The number of datapoints, classwise split of output data is:-\nclass-1 (YES)-{Y}\nclass-2 (NO)-{N}\n")
print(f"Data Summary-\n{df.describe()}")
```
